```python
# Py program to draw the activation functions
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Binary activation function.
# Returns 0 if input is less than 0, other wise 1
```

```python
np.heaviside(-10, 10)
```

```python
def binary_fun(x):
    return np.heaviside(x , 1)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, binary_fun(x))
plt.axis('tight')
plt.title('Binary activation function')
plt.show()
```

```python
# Linear activation function
# Returns what the input is
```

```python
def linear_fun(x):
    return x
```

```python
lf = np.linspace(-10,10)
plt.plot(lf, linear_fun(lf))
plt.axis('tight')
plt.title('Linear activation function')
plt.show()
```

```python
# Sigmoid function
# Returns value btn 0 & 1
# 1 / 1 + e ^ -x
#  e = np.exp
```

```python
def sigmoid_fun(x):
    return 1 / (1 +np.exp(-x))
```

```python
x = np.linspace(-10, 10)
plt.plot(x, sigmoid_fun(x))
plt.axis('tight')
plt.title("Sigmoid Function ")
plt.show()
```

```python
# Tanh activation function
# np.tanh
# (1 - e ^ - 2x) / (1 + e ^ -2x)
```

```python
def tanh_fun(x):
    return np.tanh(x)
```

```python
tf = np.linspace(-10, 10)
plt.plot(tf, tanh_fun(tf))
plt.axis('tight')
plt.title('Tanh Activation Function')
plt.show()
```

```python
# Relu 
# Return 0, if input less than 0 otherwise returns input
```

```python
def relu_fun(x):
    x1 = []
    for i in x:
        if i > 0:
            x1.append(i)
        else:
            x1.append(0)
    return x1
```

```python
rf = np.linspace(-10, 10)
plt.plot(rf, relu_fun(rf))
plt.axis('tight')
plt.title('Relu Activation Function')
plt.show()
```
